```python
in_file = open("/home/eeb177-student/Desktop/eeb-177/class-assignments/classwork-21-feb/cocoli.txt")
in_file.readline() #getting rid of the header
sizes_94 = {}
sizes_97 = {}
sizes_98 = {}

for line in in_file:
    items = line.split()
    tree_id = items[0]
    size_in_94 = float(items[4]) #makes the data sets into integers since the values have quotations around it
    size_in_97 = float(items[5])
    size_in_98 = float(items[6])
    
    sizes_94[tree_id] = size_in_94
    sizes_97[tree_id] = size_in_97
```

```python
import matplotlib.pyplot as plt

#plot some histograms
sizes_in_94 = list(sizes_94.values())
plt.hist(sizes_in_94, log = True) # log = True fixes the values of the y-axis
plt.show()
```

```python
#shows 2 graphs side by side, unlogged and logged
#very first bar is inflated due to negative values
f, (ax1, ax2) = plt.subplots(1, 2, sharex = True)
ax1.hist(sizes_in_94, log = False)
ax1.set_title("unlogged")
ax2.hist(sizes_in_94, log = True)
ax2.set_title("logged")
plt.show()
```
